```python
import index as ind
import folium
```

### renklendirmek için yazılmış fonksiyon

```python
def colorize(size):
    color=""
    if size<2:
    	return "beige"
    elif size<3.5:
        return "orange"
    elif size<4.5:
        return "red"
    elif size<5.5:
    	return "darkred"
    else:
        return "black"
       
```

# Son 500 depremi gösteren harita:

```python
m=folium.Map(
 location=[39.822060,34.808132],#middle point of turkey
 zoom_start=6)

data=ind.get_kandilli_data()

for i in range(500):
    eq=data[i]
    size=eq.get("size").get("ml")
    time=eq.get("date")
    latitude=eq.get("latitude")
    longitude=eq.get("longitude")
    depth=eq.get("depth")
    location=eq.get("location")
    folium.Marker(
    location=[latitude, longitude],
    popup=" Zaman: {} \n Yer: {} \n Büyüklük: {} ".format(time,location,size),
    icon=folium.Icon(color=colorize(size))
    ).add_to(m)
m
```
